```python
# Q1. Creating a list of 10 numbers
# These could represent any measurement — rainfall, load, etc.
numbers = [12, 45, 33, 21, 58, 67, 29, 90, 14, 39]

# The sum() function adds all values in the list
total = sum(numbers)

# len() tells us how many elements are in the list
count = len(numbers)

# Compute the average
average = total / count

average
```

```python
# Q2. Dictionary containing plant metadata
plant_info = {
    "name": "Maithon",
    "capacity_MW": 60,
    "state": "Jharkhand"
}

# .keys() returns a special object containing dictionary keys
# Wrapping it inside list() converts it into a normal Python list
keys_list = list(plant_info.keys())

keys_list
```

```python
# Q3. A list of rainfall values (in mm) for several days
rainfall = [10, 20, 5, 0, 50, 33, 12, 45, 3, 28]

# We count days where rainfall exceeded 25mm.
# A generator expression (1 for r in rainfall if r > 25)
# produces a 1 each time the condition is true.
# sum() adds all the 1's, giving us the count of such days.
count_heavy_rain = sum(1 for r in rainfall if r > 25)

count_heavy_rain
```

```python
# Q4. Function to convert Celsius to Fahrenheit
# Formula: F = (C × 9/5) + 32
def c_to_f(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Testing the function with 25°C
c_to_f(25)
```
